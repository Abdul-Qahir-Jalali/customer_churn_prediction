```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
path='/content/Churn_Modelling.csv'
data=pd.read_csv(path)
```

```python
data.head()
```

```python
data.dropna(inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.drop(columns=['RowNumber','Surname','CustomerId'],axis=1,inplace=True)
```

```python
data.head()
```

```python
x=data.drop('Exited',axis=1)
y=data['Exited']
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
scaler=StandardScaler()
label=LabelEncoder()
onehot=OneHotEncoder()
```

```python
x_copy=x.copy()
x_copy.head()
```

```python
continous_features=['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']
x_copy[continous_features]=scaler.fit_transform(x_copy[continous_features])
non_continous_features=['Geography','Gender']
x_copy[non_continous_features]=x_copy[non_continous_features].apply(label.fit_transform)
x=x_copy
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
import tensorflow
from keras.layers import Dense
from keras.models import Sequential

# ... (previous code) ...

# Define the model with the correct input_dim
ann = Sequential()
ann.add(Dense(6, input_dim=10, activation='relu')) # Changed input_dim to 10
ann.add(Dense(4, activation='relu'))
ann.add(Dense(2, activation='relu'))
ann.add(Dense(1, activation='sigmoid'))
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# ... (rest of your code) ...
```

```python
ann.fit(x_train,y_train,batch_size=100,epochs=50)
```

```python
prd=ann.predict(x_test)
```

```python
prd_data=[]
for i in prd:
  if i[0]>0.5:
    prd_data.append(1)
  else:
    prd_data.append(0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prd_data)
```

```python
# Evaluate accuracy on training data
train_loss, train_accuracy = ann.evaluate(x_train, y_train)

# Evaluate accuracy on testing data
test_loss, test_accuracy = ann.evaluate(x_test, y_test)

print(f"Training Accuracy: {train_accuracy * 100:.2f}%")
print(f"Testing Accuracy: {test_accuracy * 100:.2f}%")
```
